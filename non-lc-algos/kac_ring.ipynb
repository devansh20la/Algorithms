# Kac Ring (Interview)

**Topics:** Simulation, Array  
**Origin:** Mark Kac’s ring model (1956).

---

## Description

There are **N** balls on sites `0, 1, …, N - 1` around a **circle**. Each ball is **white** (`0`) or **black** (`1`).

A list **`marked_points`** holds distinct site indices in `[0, N - 1]`. A mark on site `i` places a fixed flipper on the edge **i → (i + 1) mod N**.

One **global step** (all balls move at once):

1. Every ball at site `i` moves **clockwise** to `(i + 1) mod N`.
2. If site `i` is marked, the ball **flips color** on that move.

---

## Your task

Implement class **`KacRing`** in the next cell with **only** these two methods:

### `color(N)`

Return the **initial** colors of all `N` balls. Every ball starts **white**.

- **Input:** `N` (number of sites)
- **Output:** length-`N` list of ints in `{0, 1}`

### `step(colors, marked_points)`

Apply **one** Kac-ring step and return the new colors.

- **Input:** current `colors` (length `N`), and `marked_points`
- **Output:** length-`N` list after one simultaneous rotate-and-flip step

Do **not** change the class name, method names, or signatures.

## Examples

**Example 1**

```python
ring = KacRing()
ring.color(4)                      # -> [0, 0, 0, 0]
ring.step([0,0,0,0], [0, 2])       # -> [0, 1, 0, 1]
```

Balls at sites `0` and `2` cross flippers and arrive black at sites `1` and `3`.

**Example 2**

```python
ring = KacRing()
c = ring.color(5)
ring.step(c, [])              # -> [0, 0, 0, 0, 0]  (pure rotation of white)
```

**Example 3**

```python
ring = KacRing()
c = ring.color(3)
c = ring.step(c, [0, 1, 2])   # -> [1, 1, 1]
ring.step(c, [0, 1, 2])       # -> [0, 0, 0]
```

## Tests

Implement `KacRing` above, then run:

run_tests()

````python
from typing import List


def black_count(colors: List[int]) -> int:
    return sum(colors)


def run_after_t(ring: KacRing, N: int, marked_points: List[int], t: int) -> List[int]:
    c = ring.color(N)
    for _ in range(t):
        c = ring.step(c, marked_points)
    return c


def run_tests():
    ring = KacRing()

    assert ring.color(4) == [0, 0, 0, 0]
    assert ring.step([0, 0, 0, 0], [0, 2]) == [0, 1, 0, 1]

    assert ring.step(ring.color(5), []) == [0, 0, 0, 0, 0]

    c = ring.color(3)
    c = ring.step(c, [0, 1, 2])
    assert c == [1, 1, 1]
    assert ring.step(c, [0, 1, 2]) == [0, 0, 0]

    assert black_count(run_after_t(ring, 4, [0, 2], 1)) == 2
    assert black_count(run_after_t(ring, 5, [], 100)) == 0
    assert black_count(run_after_t(ring, 3, [0, 1, 2], 2)) == 0
    assert black_count(run_after_t(ring, 2, [0, 1], 2)) == 0

    print("All tests passed.")
````
