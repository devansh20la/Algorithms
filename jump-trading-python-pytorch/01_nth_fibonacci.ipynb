## Interview Prompt

Write `fib(n)` that returns the nth Fibonacci number where `fib(0)=0` and `fib(1)=1`. Your solution should handle large `n` without exponential recursion.


## Constraints

- 0 <= n <= 10_000
- Return the exact integer, not a floating point approximation
- Target O(n) time and O(1) extra space, then discuss faster methods


## Examples

```python
fib(0) -> 0
fib(1) -> 1
fib(10) -> 55
fib(50) -> 12586269025
```


## Starter Code

Implement the function or class below. Keep your first pass simple and correct, then discuss complexity and edge cases out loud as you would in an interview.


````python
from functools import lru_cache

@lru_cache(maxsize=None)
def fib(n: int) -> int:
    """Return the nth Fibonacci number."""
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fib(n - 1) + fib(n - 2)
````

## Sample Tests

Run this cell after implementing your solution. Add your own edge cases before considering the problem complete.


````python
assert fib(0) == 0
assert fib(1) == 1
assert fib(2) == 1
assert fib(10) == 55
assert fib(50) == 12586269025
print('sample tests passed')
````
