## Interview Prompt

Implement a small limit order book that supports adding limit orders, canceling orders, and returning the current best bid/ask. For this practice version, orders do not match; they just rest on the book.


## Constraints

- Each order has `order_id`, `side`, `price`, and `quantity`
- `side` is `'B'` for bid or `'A'` for ask
- Canceling a missing order should return `False`
- Best bid is highest bid price; best ask is lowest ask price
- Multiple orders can share the same price


## Examples

```python
book.add('o1', 'B', 100.5, 10)
book.add('o2', 'A', 101.0, 3)

```
