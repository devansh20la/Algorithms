## Interview Prompt

Implement `running_mean_std(values)` that consumes an iterable of numbers once and returns `(mean, population_std)`. It should be numerically stable for large values.


## Constraints

- Input can be any iterable, including a generator
- Do not materialize the whole input
- Use population standard deviation
- Raise `ValueError` for an empty input


## Examples

```python
running_mean_std([2, 4, 4, 4, 5, 5, 7, 9]) -> (5.0, 2.0)
```

````python
from collections.abc import Iterable
import math 

def running_mean_std(values: Iterable[float]) -> tuple[float, float]:
    """Return mean and population standard deviation in one pass."""
    mean = 0.0
    std = 0.0
    count = 0
    for v in values:
        mean_sum += v
        sq_sum += v**2
        count += 1
        
        mean = mean_sum / count
        var = sq_sum / count - mean**2
        yield mean, math.sqrt(var)
    
````
