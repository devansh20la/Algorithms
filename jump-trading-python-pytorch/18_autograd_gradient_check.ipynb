````python
import torch

def f(z):
    return (z ** 2).sum() + 3 * z[0]

x = torch.tensor([1.0, -2.0, 0.5], requires_grad=True)
y = f(x)
y.backward()
num_grad = finite_difference_grad(f, x.detach())
assert num_grad.shape == x.shape
assert torch.allclose(num_grad, x.grad, atol=1e-3, rtol=1e-3)
print('sample tests passed')
````
