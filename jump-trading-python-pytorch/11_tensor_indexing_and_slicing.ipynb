## Interview Prompt

Implement `select_token_features(x, batch_idx, token_idx)` for a 3D tensor `x` with shape `(batch, seq_len, hidden)`. Return the hidden vectors at `(batch_idx[i], token_idx[i])` for every `i` without using a Python loop.


## PyTorch Concepts Tested

- Advanced integer indexing
- Shape reasoning
- Avoiding Python loops over tensor rows


## Constraints

- `x` is a `torch.Tensor` with rank 3
- `batch_idx` and `token_idx` are 1D `LongTensor`s with the same shape
- Return shape must be `(len(batch_idx), hidden)`
- Preserve dtype and device of `x`


## Examples

```python
x.shape == (2, 3, 4)
batch_idx = torch.tensor([0, 1])
token_idx = torch.tensor([2, 0])
select_token_features(x, batch_idx, token_idx).shape == (2, 4)
```


## Starter Code

Fill in the implementation. As in a real interview, explain tensor shapes at every step before optimizing.


````python
import torch

def select_token_features(x: torch.Tensor, batch_idx: torch.Tensor, token_idx: torch.Tensor) -> torch.Tensor:
    return x[batch_idx, token_idx, :]
````

## Sample Tests

Run this cell after implementing your solution. Add extra edge cases, especially around shape, dtype, and device behavior.


````python
import torch

x = torch.arange(2 * 3 * 4).reshape(2, 3, 4)
batch_idx = torch.tensor([0, 1, 1])
token_idx = torch.tensor([2, 0, 2])
out = select_token_features(x, batch_idx, token_idx)
expected = torch.stack([x[0, 2], x[1, 0], x[1, 2]])
assert torch.equal(out, expected)
assert out.shape == (3, 4)
print('sample tests passed')
````
