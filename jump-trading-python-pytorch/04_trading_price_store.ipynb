## Interview Prompt

Design `PriceStore`, a class for storing latest prices by symbol. It should support setting a price, getting one price, and returning the top-k symbols by latest price.


## Constraints

- Symbol is a non-empty string
- Price must be a positive float
- Getting a missing symbol should raise `KeyError`
- `top_k(k)` returns symbols sorted by price descending, then symbol ascending for ties


## Examples

```python
store.set_price('ES', 5025.25)
store.get_price('ES') -> 5025.25
store.top_k(1) -> [('ES', 5025.25)]
```
