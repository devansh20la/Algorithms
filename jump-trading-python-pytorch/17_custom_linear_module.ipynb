## Sample Tests

````python
import torch

layer = MyLinear(4, 2)
x = torch.randn(3, 4)
y = layer(x)
assert y.shape == (3, 2)
assert dict(layer.named_parameters())['weight'].shape == (2, 4)
assert dict(layer.named_parameters())['bias'].shape == (2,)

no_bias = MyLinear(4, 2, bias=False)
assert 'bias' not in dict(no_bias.named_parameters())
assert no_bias(torch.randn(5, 7, 4)).shape == (5, 7, 2)
print('sample tests passed')
````
